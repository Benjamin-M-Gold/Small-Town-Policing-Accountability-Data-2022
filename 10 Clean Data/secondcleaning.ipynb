```python
#looking through all columns to see if there are NA values
arrests["race"].isna().sum()

#if yes, looking at the rows with NA values to see what's up
arrests.loc[arrests["race"].isna()]

#looking at rows where both race and sex are NA
arrests[(arrests['race'].isna()) & (arrests['sex'].isna())]
```

```python
#investigating the number of each race entered, looked deeper into race "O"
arrests["race"].value_counts()
arrests.loc[arrests["race"] == "O"]
```

```python
#separating datetime columns
arrests['dates'] = pd.to_datetime(arrests['datetimeofarrest']).dt.date
arrests['time'] = pd.to_datetime(arrests['datetimeofarrest']).dt.time
arrests['year'] = pd.to_datetime(arrests['datetimeofarrest']).dt.year
arrests['month'] = pd.to_datetime(arrests['datetimeofarrest']).dt.month
arrests['day'] = pd.to_datetime(arrests['datetimeofarrest']).dt.day
```

```python
#dropping all rows with duplicate values for arrest number (same pdf just entered twice)
# saves a copy as a new object

arrests = arrests.drop_duplicates(['arrestnumber', 'dates']).copy()
```

```python
#looking into value counts of different variables to see what needs cleaning
arrests["race"].value_counts()
```

```python
#replacing all weird values for eye color
arrests.loc[arrests["race"] == "O", "race"] = "U"

arrests.loc[arrests["eyecolor"] == "Brow", "eyecolor"] = "Brown"
arrests.loc[arrests["eyecolor"] == "Br", "eyecolor"] = "Brown"
arrests.loc[arrests["eyecolor"] == "Blac", "eyecolor"] = "Black"
arrests.loc[arrests["eyecolor"] == "Hz", "eyecolor"] = "Hazel"
arrests.loc[arrests["eyecolor"] == "Be", "eyecolor"] = "Brown"
arrests.loc[arrests["eyecolor"] == "Other", "eyecolor"] = "Unknown"
```

```python
#replacing all weird values for hair color
arrests.loc[arrests["haircolor"] == "Blo", "haircolor"] = "Blonde"
arrests.loc[arrests["haircolor"] == "Bln", "haircolor"] = "Blonde"
arrests.loc[arrests["haircolor"] == "Bld", "haircolor"] = "Unknown/Bald"
arrests.loc[arrests["haircolor"] == "Gy", "haircolor"] = "Gray"
arrests.loc[arrests["haircolor"] == "Bk", "haircolor"] = "Black"
arrests.loc[arrests["haircolor"] == "Br", "haircolor"] = "Brown"
arrests.loc[arrests["haircolor"] == "Xxx", "haircolor"] = "Unknown/Bald"
```
